## Выполнение SQL-запросов
### Количество книг, которые вышли после 1 января 2000 года
**Вывод:** 
- После 1 января 2000 года вышло 819 книг. 

```python
# чтобы выполнить SQL-запрос, используем Pandas
query = '''
           SELECT count(book_id) "Количество книг"
           FROM books 
           WHERE publication_date > '2000-01-01' 
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### Количество обзоров и средняя оценка для каждой книги
**Вывод:** 
- Всего в базе представлено 1000 книг. 
- Самое большое количество обзоров написано для книги Twilight (7 обзоров). Ее средняя оценка - 3.66.
- Далее идут 14 книг с количеством обзоров равным 6. Среди них книги с самым высоким рейтингом это:
    - Harry Potter and the Prisoner of Azkaban — 4.41
    - Harry Potter and the Chamber of Secrets — 4.29
    - The Book Thief — 4.26

```python
# чтобы выполнить SQL-запрос, используем Pandas
query = '''
           SELECT b.title "Название книги",
                  COUNT(DISTINCT rev.review_id) "Количество обзоров", 
                  ROUND(AVG(rat.rating), 2) "Средняя оценка"
           FROM books b
           LEFT JOIN ratings rat ON b.book_id=rat.book_id 
           LEFT JOIN reviews rev ON b.book_id=rev.book_id
           GROUP BY b.book_id
           ORDER BY "Количество обзоров" DESC, "Средняя оценка" DESC
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### Издательство, которое выпустило наибольшее число книг толще 50 страниц
**Вывод:** 
- Издательство, выпустившее наибольшее число книг толще 50 страниц — Penguin Books. Это издательство выпустило 42 книги.

```python
# чтобы выполнить SQL-запрос, используем Pandas
query = '''
           SELECT p.publisher "Издательство",
                  COUNT(DISTINCT b.book_id) "Количество книг"
           FROM publishers p
           LEFT JOIN books b ON b.publisher_id=p.publisher_id
           WHERE b.num_pages > 50
           GROUP BY p.publisher
           ORDER BY "Количество книг" DESC
           LIMIT 1
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### Автор с самой высокой средней оценкой книг для книг с 50 и более оценками
**Вывод:**  
- Автор с самой высокой средней оценкой книг для книг с 50 и более оценками — J.K. Rowling/Mary GrandPré. Средняя оценка автора — 4.29.

```python
# чтобы выполнить SQL-запрос, используем Pandas
query = '''
           SELECT a.author "Автор", ROUND(AVG(r.rating), 2) "Средняя оценка"
           FROM authors a
           JOIN books b ON a.author_id=b.author_id 
           AND b.book_id IN (SELECT book_id
                             FROM ratings
                             GROUP BY book_id
                             HAVING COUNT(DISTINCT rating_id) >= 50)
           LEFT JOIN ratings r ON b.book_id=r.book_id
           GROUP BY a.author_id
           ORDER BY "Средняя оценка" DESC
           LIMIT 1
                      
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

### Среднее количество обзоров от пользователей, которые поставили больше 48 оценок
**Вывод:** 
- Среднее количество обзоров от пользователей, которые поставили больше 48 оценок — 24.

```python
# чтобы выполнить SQL-запрос, используем Pandas
query = '''
           SELECT AVG(count):: int "Количество обзоров"
           FROM (SELECT username, COUNT(review_id)
                 FROM reviews
                 WHERE username IN (SELECT username
                                    FROM ratings
                                    GROUP BY username
                                    HAVING COUNT(rating) > 48) 
                 GROUP BY username) t
        '''

con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```
